```python
import random

def generate_device_info():
    android_versions = [10, 12, 13, 14, 15, 16]
    chrome_versions = ["144.0.0.0", "145.0.7632.120", "146.0.0.0", "146.0.7680.177", "147.0.0.0"]
    
    if random.random() < 0.75:  # mostly Android
        android = random.choice(android_versions)
        chrome = random.choice(chrome_versions)
        
        return f"Mozilla/5.0 (Linux; Android {android}; K) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/{chrome} Mobile Safari/537.36"
    
    else:  # iPhone
        ios_versions = ["18_7", "18_3_1", "17_6"]
        ios = random.choice(ios_versions)
        
        return f"Mozilla/5.0 (iPhone; CPU iPhone OS {ios} like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/26.3 Mobile/15E148 Safari/604.1"
```

```python
df['device_info'] = df['device_info'].apply(
    lambda x: generate_device_info() if pd.isna(x) else x
)
```
